```python
#question 3 C#

using System;

public class area
{
    public static double triangleArea(double a, double b, double c)
    {
        double s = (a + b + c) / 2;

        double totalArea = Math.Sqrt(s * (s - a) * (s - b) * (s - c));
        return totalArea;
    }

    public static void Main()
    {
        Console.Write(" Input the lengths of the 3 sides of a triangle: ");

        double a = Convert.ToDouble(Console.ReadLine());

        double b = Convert.ToDouble(Console.ReadLine());

        double c = Convert.ToDouble(Console.ReadLine());

        Console.WriteLine("\nThe area of the triangle : {0} \n", triangleArea(a, b, c));

        Console.ReadLine();
    }
}
```

```python
#question 5 C#

using System;

class question5
{

 
    static bool inside(double x1, double y1, double x2,
                          double y2, double x, double y)
    {
        if (x > x1 && x < x2 &&
            y > y1 && y < y2)
            return true;

        return false;
    }

    public static void Main()
    {

        Console.Write("Enter points(x, y):");
        double x=double.Parse(Console.ReadLine());
        double y=double.Parse(Console.ReadLine());
        Console.Write("Enter corners of rectangle(x1, y1, x2, y2):");
        double x1=double.Parse(Console.ReadLine());
        double y1=double.Parse(Console.ReadLine());
        double x2=double.Parse(Console.ReadLine());
        double y2=double.Parse(Console.ReadLine());

        if (inside(x1, y1, x2, y2, x, y))
            Console.Write("True");
        else
            Console.Write("False");
    }
}
```

```python
#question 6

def pig(word):  # Defining pig () function and passing word parameter
    len_of_word = len(word)  # Assigning length of word in variable
    lower_case_word = word.lower()  # Converting letters in lower case 
    l1=['a','e','i','o','u']
    if lower_case_word in l1:
        lower_case_word+='way'  # Adding 'way' string into lower case word into vowels
    else:
        lower_case_word = lower_case_word[1:len_of_word]+lower_case_word[0]+'ay'  # For consonants it will print other letter except first one and then first letter and at last it will add 'ay'
    return lower_case_word
print(pig('aappy'))
print(pig('Enter'))

  



```

```python
#question 7

def bldcount():  # Defining bldcount functon 
    file_path=r"bloodtype1.txt"    # Reading the text file
    fp = open(file_path, "r")      
    people_count = fp.read().split(' ')  # Spliting the content
    print('There are {} patients of blood type A.'.format(people_count.count('A')))    # Counting Blood group A from file
    print('There are {} patients of blood type B.'.format(people_count.count('B')))    # Counting Blood group B from file
    print('There are {} patients of blood type AB.'.format(people_count.count('AB')))  # Counting Blood group AB from file
    print('There are {} patients of blood type O.'.format(people_count.count('O')))    # Counting Blood group o from file
    print('There are {} patients of blood type OO.'.format(people_count.count('OO')))  # Counting Blood group oo from file
    
bldcount()
```
